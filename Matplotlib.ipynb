Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[10,15,16,17,20,25,30]
y=[50,52,55,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='green',linewidth=5, linestyle='dotted')
plt.savefig("Image01.png")
```

```python
plt.plot(x,y, 'g+')
plt.savefig("Image02.png")
```

```python
plt.plot(x,y, 'g+-.')
plt.savefig("Image03.png")
```

```python
plt.plot(x,y, '--r*')
plt.savefig("Image04.png")
```

```python
plt.plot(x,y, 'rD')
plt.savefig("Image05.png")
```

```python
plt.plot(x,y,color='red',marker='D', linestyle='')
plt.savefig("Image06.png")
```

```python
plt.plot(x,y,color='green',marker='+', linestyle='', markersize=30)
plt.savefig("Image07.png")
```

```python
plt.plot(x,y,color='green', alpha=0.2)
plt.savefig("Image08.png")
```
